```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
%load_ext autoreload
%autoreload 2
from path_import import *
from render import *
```

```python
file = "../../clasical_ray_marching/paths.json"
pixels = parse_file(file)
```

```python
(rows, cols) = (5,5)
result_image_r = [[(1,1) for j in range(cols)] for i in range(rows)]
result_image_g = [[(1,1) for j in range(cols)] for i in range(rows)]
result_image_b = [[(1,1) for j in range(cols)] for i in range(rows)]
result_images = [result_image_r, result_image_g, result_image_b]
```

```python

filter = True
start = [4,3]
end = [4,4]
if filter:
    to_render = []
    for i in range(start[0], end[0] +1):
        for j in range(start[1], end[1] +1):
            for c in range(3):
                to_render.append([i,j, c])
    render(pixels, result_images, filter = to_render)
else:
    render(pixels, result_images)


value_image = np.zeros((rows,cols))
r = np.zeros((rows,cols))
g = np.zeros((rows,cols))
b = np.zeros((rows,cols))

qr = np.zeros((rows,cols))
qg = np.zeros((rows,cols))
qb = np.zeros((rows,cols))
for pi in pixels:
    x = pi.pixel[0]
    y = pi.pixel[1]
    r[y][x] = result_images[0][x][y][1]
    g[y][x] = result_images[1][x][y][1]
    b[y][x] = result_images[2][x][y][1]

    qr[y][x] = 0 if result_images[0][x][y] is None else result_images[0][x][y][0]
    qg[y][x] = 0 if result_images[1][x][y] is  None else result_images[1][x][y][0]
    qb[y][x] = 0 if result_images[2][x][y] is  None else result_images[2][x][y][0]


actuall_image = (np.dstack((r,g,b)).clip(0.0,1.0) * 255).astype(np.uint8)
# print(actuall_image)
q_image = (np.dstack((qr,qg,qb)).clip(0.0,1.0)  * 255) .astype(np.uint8)

gs = gridspec.GridSpec(1, 3)

plt.figure()
ax = plt.subplot(gs[0, 0]) # row 0, col 0
plt.axis('off')
plt.imshow(actuall_image)

ax = plt.subplot(gs[0, 1]) # row 0, col 1
plt.axis('off')
plt.imshow(q_image)

ax = plt.subplot(gs[0, 2]) # row 1, span all columns
plt.axis('off')
plt.imshow(abs(actuall_image - q_image))
plt.savefig("combined_image.png", bbox_inches='tight')
```
